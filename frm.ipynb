```python
ramp1 = np.linspace(-3,3,6000)
ramp2 = np.linspace(-3,3,6000)
ramp3 = np.linspace(-3,3,6000)
ramp = np.concatenate((ramp1,ramp2,ramp3))
t = np.linspace(-0.25,0.65,18000)
```

```python
plt.plot(t,ramp, label='flux ramp')
plt.plot(t,ramp+0.2, label='offset flux ramp')
plt.xlim(0,0.4)
plt.ylabel('$\Phi_{\mathrm{mod}}$ / $\Phi_0}$')
plt.xlabel('t / a.u.')
plt.legend()
```

```python
plt.plot(t,0.8+0.3*np.sin(ramp*6))
plt.plot(t,0.8+0.3*np.sin(ramp*6+np.pi/2))
plt.xlim(0,0.4)
plt.ylim(0,1.5)
plt.ylabel('$V(t)$ / a.u.')
plt.xlabel('t / a.u.')
```

```python
plt.plot(t,1.2+0.3*np.sin(ramp*6+np.pi/2)+0.3*np.sin(ramp*10))
plt.xlim(0,0.4)
plt.ylim(0,2.5)
plt.ylabel('$V_{12}(t)$ / a.u.')
plt.xlabel('t / a.u.')
```

```python
plt.plot(t,0.6+0.3*np.sin(ramp*10))
plt.xlim(0,0.4)
plt.ylim(0,1.25)
plt.ylabel('$V_1(t)$ / a.u.')
plt.xlabel('t / a.u.')
```
