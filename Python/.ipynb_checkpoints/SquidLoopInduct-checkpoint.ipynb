```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('genplot_py3.mplstyle')
```

```python
mu_0 = 4*np.pi*1e-7 # H/m

#Calculate Inductance of polygonal loop
#Variables: 
#N = number of corners
#d = diameter of inner polygon
#w = width of polygonal loop

# c describes the hole perimeter
def c(N,d):
    return N*np.tan(np.pi/N)*d

# geometric constant of inductance
def gamma_L(d,w,N):
    return (np.log(d/w+2.8)+0.2+2.7*d/c(N,d))/2.07

# inductance of polygonal loop 
def L_h(d,w,N):
    return gamma_L(d,w,N)*mu_0*c(N,d)/np.pi
```

```python
width_array = np.logspace(0,3,100)*1e-6
L_h_array1 = np.array([])
L_h_array2 = np.array([])
for w in width_array:
    L_h_array1 = np.append(L_h_array1,L_h(100e-6,w,8))
    L_h_array2 = np.append(L_h_array2,L_h(50e-6,w,8))
```

```python
plt.plot(width_array*1e6,L_h_array1*1e9, label = 'octagon, d = 100m')
plt.plot(width_array*1e6,L_h_array2*1e9, label = 'octagon, d = 50m')
plt.xscale('log')
plt.xlabel('washer width [m]')
plt.ylabel('washer inductance [nH]')
plt.legend()

#plt.savefig('SquidIndWidth.pdf')
plt.show()
```

```python
dia_array = np.logspace(1,2,100)*1e-6
L_h_array1 = np.array([])
L_h_array2 = np.array([])
for d in dia_array:
    L_h_array1 = np.append(L_h_array1,L_h(d,10e-6,8))
    L_h_array2 = np.append(L_h_array2,L_h(d,5e-6,8))
```

```python
plt.plot(dia_array*1e6,L_h_array1*1e9, label = 'octagon, w = 10m')
plt.plot(dia_array*1e6,L_h_array2*1e9, label = 'octagon, w = 5m')
#plt.xscale('log')
plt.xlabel('washer loop diameter [m]')
plt.ylabel('washer inductance [nH]')
plt.legend()

#plt.savefig('SquidIndDia.pdf')
plt.show()
```

```python
def d(w,l,N):
    return l/(N*np.tan(np.pi/N))-w

w_array = np.array([1.4,2.1,3.5])*1e-6
l_array = np.array([350,700,1400,2100])*1e-6


L_h_array = np.zeros([3,4])
for i,w in enumerate(w_array):
    for j,l in enumerate(l_array):
        d_cur = d(w,l,8)
        L_h_array[i][j] = L_h(d_cur,w,8)
```
