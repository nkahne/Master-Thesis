```python
import matplotlib.pyplot as plt
import numpy as np
from itertools import cycle
from matplotlib.cm import ScalarMappable

plt.rcParams['figure.figsize'] = [12, 12]
plt.rcParams['legend.fontsize'] = 25
plt.rcParams['axes.labelsize'] = 35
plt.rcParams['font.size'] = 30
plt.rcParams['xtick.labelsize'] = 20
plt.rcParams['ytick.labelsize'] = 20

# Sample data
x = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
y = np.arange(1, 53)
z = [
    np.random.randint(1, 20, 52),
    np.random.randint(1, 20, 52),
    np.random.randint(1, 20, 52),
    np.random.randint(1, 20, 52),
    np.random.randint(1, 20, 52)
]

# Check if the dimensions of z match the dimensions of x and y
if len(x) != len(z) or len(y) != len(z[0]):
    raise ValueError("Dimensions of z do not match x and y axes.")

# Create a figure and a 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Define colors for each weekday
colors = cycle(['red', 'green', 'blue', 'orange', 'purple'])

# Calculate normalized colors based on height values
norm = plt.Normalize(np.min(z), np.max(z))
cmap = plt.get_cmap('viridis')
scalar_map = ScalarMappable(norm=norm, cmap=cmap)

# Iterate over the data and plot the stems
for i in range(len(x)):
    color = next(colors)
    for j in range(len(y)):
        stem_color = scalar_map.to_rgba(z[i][j])
        ax.plot([i, i], [y[j], y[j]], [0, z[i][j]], color=stem_color)

# Set labels and title
ax.set_xlabel('Day of the week')
ax.set_ylabel('Calendar week')
ax.set_zlabel('Value')
ax.set_xticks(range(len(x)))
ax.set_xticklabels(x)

ax.set_title('3D Stem Plot')

# Add colorbar
scalar_map.set_array(z)  # Set the colorbar values based on z
cbar = plt.colorbar(scalar_map)
cbar.set_label('Value')

# Display the plot
plt.show()
```
