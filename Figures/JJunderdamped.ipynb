```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [10, 10]
plt.rcParams['legend.fontsize'] = 15
plt.rcParams['axes.labelsize'] = 25
plt.rcParams['font.size'] = 25
plt.rcParams['xtick.labelsize'] = 20
plt.rcParams['ytick.labelsize'] = 20
plt.rcParams['axes.titlesize'] = 20
plt.rcParams['axes.prop_cycle'] = plt.cycler(color=['#1b85b8','#ae5a41','#559e83','#c3cb71','#fb9062'])
plt.rcParams['ytick.direction'] = 'in'
plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['legend.frameon'] = False
```

```python
gen = np.genfromtxt('9_gen.csv', skip_header=2, delimiter=",")
sig = np.genfromtxt('9_sig.csv', skip_header=2, delimiter=",")

gen_t = gen.T[3]
gen_sig = gen.T[4]

sig_t = sig.T[3]
sig_sig = sig.T[4]
```

```python
f, ax = plt.subplots()
Ic=0.096
R=82.235
IcR = Ic*R*1e-1
ax.plot(sig_sig*10,gen_sig/Ic, label='HDSQ15w1, chip 2A19')
ax.set_ylim(min(gen_sig/Ic),max(gen_sig/Ic))
ax.set_xlim(min(sig_sig*10),max(sig_sig*10))
ax.set_xlabel(r'average voltage $V$ [mV]')
ax.set_ylabel(r'current $I/I_{\rm c}$ ')
ax.text(0.97,0.02,r'$I_{\rm c}=9.6{\rm µA}$', ha='right', va='bottom', color = 'grey', transform=ax.transAxes)
ax.legend()
plt.savefig('HDSQ15w1_2A19_9_windowJJ_IVC.pdf')
plt.show()
```
