```python
x = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
y = np.arange(1, 71)
# Transpose the z array
z = np.array(z).T

# Create a 2D histogram using imshow
plt.imshow(z, cmap='viridis', origin='lower', aspect='auto')

# Set x-axis ticks and labels
plt.xticks(range(len(x)), x)

# Set y-axis ticks and labels
plt.yticks(range(len(y)), y)

# Set labels and title
#plt.xlabel('Day of the week')
plt.ylabel('Calendar week')
plt.title('Coffee Corner Attendance')

#yticks = np.arange(0, 68, 10)
#plt.yticks(yticks, y[yticks])

yticks = np.arange(0, 71, 10)
plt.yticks(yticks)
yticks_labels = [str(i) for i in yticks]
yticks_labels[0] = '0'
plt.gca().yaxis.set_major_locator(ticker.FixedLocator(yticks))
plt.gca().yaxis.set_major_formatter(ticker.FixedFormatter(yticks_labels))

mean=[np.mean(i) for i in z.T]
rounded = [round(i,2) for i in mean]
total=[np.sum(i) for i in z.T]

specific_info = rounded 
additional_info = total
line_height = 0
for i, label in enumerate(x):
    plt.text(i, line_height - 5, specific_info[i], ha='center', va='top',fontsize=20)
    plt.text(i, line_height - 10, additional_info[i], ha='center', va='top',fontsize=20) 

plt.text(-0.6,line_height - 5,r'$\bar{x}$:',ha='center', va='top',fontsize=20)
plt.text(-0.6,line_height - 10,r'$\Sigma$:',ha='center', va='top',fontsize=20)

#Display the plot
plt.colorbar()
plt.savefig('CoffeeCornerPlot3b.pdf')
plt.show()
```
