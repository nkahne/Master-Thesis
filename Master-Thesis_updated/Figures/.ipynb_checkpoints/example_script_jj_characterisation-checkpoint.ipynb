# Examplary Layout to evaluate JJ IV-Curves

This script is supposed to act as an example on how the squid_gamma library may be used to automate the characterisation of Josephson Junctions.

```python
import numpy as np   # typical import of useful modules
import matplotlib.pyplot as plt
import squid_delta as squid # the squid library containes the functions to evaluate JJ's

%matplotlib inline 
# use the 'inline' backend to get fast and clean output of plots.
# use 'tk' to get interactive plots in seperate windows (gets cluttered for large numbers of plots)
```

The 'squid_gamma' library contains the functions to automatically characterise a Josephson Junction from its IV Curve. In the cell below, the folder location conataining the data (pairs of _gen.csv and _sig.csv files) is defined, and the name of the .txt file containing the obtained junction parameters is specified.

Note that the script will automatically find all file pairs in a given folder, so the location should specify a folder, not an individual file.

```python
file1 = '\\\\twix2\\bolo\\data\\JosephsonJunctions\\HDSQ15\\HDSQ15w1\\Chip_2A19\\' # folder locations we want to use
fname1 = 'Results_SingleJunctions_2A19' # name for the .txt file containing all the data
```

# CrossJJ4_w8

The 'squid.findfiles' function takes a folder location and searches for mathing pairs of _gen.csv and _sig.csv files. (If for some reason, different identifiers than _gen and _sig have been used, the new identifiers can be passed to the function as inputs 'gen=' and 'sig=').

If only some of the files in a folder are supposed to be imported, an additional name specifier ('name=') using regular expressions can be passed to the function.

The function returns a dicitionary, where the stored x- and y-data can be obtained by passing the proper key to the dictionary. The 'squid.findfiles' function will print the names of all found pairs. These names are also the keys under which the corresponding information can be found.

```python
data1 = squid.findfiles(file1)

ends = ['11','9','7','6','5','2'] # to get the output in this specific order (channel numbers)
kt = '' # key template. This template together with the endings in the list above gives the full keys
```

The following is a very simple loop to do the same operation for all files in the previously generated dictionary. The 'squid.plot_jj' function is used, which plots the data from the files and automatically attempts to read out the critical current, retrapping current, gap voltage, normal resistance, subgap resistance and interception points (IC points). These values are printed as output, but may also be returned as a local variable (set output to 'True') or be given as a textbox within the plot itself (set textbox to 'True').

A gain of 100 and a resistance of 10.3 $k\Omega$ to create a current from the supplied voltage are set as standard. If other values were used, they can be passed to the function using the 'gain=' and 'supply_resistor=' inputs.

Since the function looks for specific features in the data, the characterisation is likely to (partially) fail for uncommonly shaped IV curves. To quickly asses the quality of the characterisation, set telemetry to 'True'. This will mark the found values or plot the fit functions into the data, so it can easily be seen if the found values agree with the data.

The function 'squid.init_file_jj' initialises a .txt file at the given path. We will later write the gathered information on the JJ's to this file. Please note that this init-function will create a new file, deleting the old file of the same name if present. The parameter 'name' simply refers to the first line in the file, not the actual file name (change this via the path)

The function 'squid.writeline_jj' simply writes the data on a JJ ('vals', output of squid.plot_jj) to the given file with a specific format.

The 'squid.writeline_jj' function will attempt to filter out bad data from unsuccesfull characterisations. If the found gap voltage differs by more than 0.3 mV from 2.7 mV, the wirtten line of the found values will be commented out in the text file. This is to prevent false informtion being read from the .txt file in later data analysis. Also, the values in the 'Size' and 'R300' columns are replaced with 'CHECK MANUALLY' to alert the user that something went wrong, and that manual characterisation of the junction may be necessary.

```python
squid.init_file_jj(file1+fname1+'.txt',comment='') # initialise a txt file to write values to
# please note that 'comment' refers to the first line in the file

#for key in data1: # this does not require any 'ends' list, but will give a random order of the output and .txt file
for end in ends:   # this will order the output (and .txt file)in the way specified in the 'ends' list
    key = kt+end # constructs the key from the template and ending. Not needed if dictionary keys are used directly
    dats = data1[key]
    plt.figure(figsize=(9,6),dpi=100) # defines plot size (in inches, sadly) and resolution (in dots per inch)
    plt.title(key)
    
    #plt.ylim((-22,22)) # place axis limits before squid.plot_jj to avoid bugs in textbox placement
    #plt.xlim((-1,1))
    
    vals = squid.plot_jj(dats,key=key,unit_func=squid.original,telemetry=True,output=True,textbox=True)
    # output order of squid.plot_jj if enabled: icp,icn,irp,irn,vgp,vgn,Rnp,Rnn,Rsgp,Rsgn,ICPp,ICPn = 
    # ending p designates positive, ending n designates negative value
    # if only a single variable is given for the output (such as 'vals'), a tuple of all values will be passed instead
    
    plt.legend()
    
    plt.savefig(file1+key+'.png',ftype='png',dpi='figure') # save plots as .eps with previously specified dpi
    plt.show()

    squid.writeline_jj(key,*vals,file1+fname1+'.txt') # append new line of values to the previously initiated file
```
